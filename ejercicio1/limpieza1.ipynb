```python
import numpy as np
```

```python
def cambiar_coma_punto(columna):
    try: 
        return float(columna.replace(",", "."))
    except:
        return np.nan
        
```

```python
lista_columnas = ["cons,price,idx", "cons,conf,idx", "euribor3m", "nr,employed"]
```

```python
for columna in lista_columnas:
    df[columna] = df[columna].apply(cambiar_coma_punto)
```
